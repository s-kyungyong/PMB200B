###For the sake of time, we will not go over what each prameter means and use default options for most of the parameters, which typically works fine. These are the parameters you need to pay attentions.

###**query_sequence**: protein sequences you want to predict  

###**num_relax**: the number of models for which you want to relax side chains (i.e. amino acid side chain optimization). In most cases, **1** works fine. This will relax the side chain for the best model

###**template_mode**: Typically, **pdb100** works fine. With this parameter on, AlphaFold will search for homologous sequences with experimentally solved structures and use the information to model your protein.

###**model_type**: For monomers, use alphafold2_ptm; for protein complexes (two or more proteins forming a complex), use alphafold2_multimer_v3. But **auto** works well and use the proper model based on the input sequences.

###Within the ColabFold notbook, you can visualize and play with the predicted structure. There are two major metrics you need to pay attention:

###**pLDDT**: Predicted local residue accuracy. In other words, how accurately each residue would be predicted. This ranges from 0 to 100 (some may use 0 to 1 scale instead). The average pLDDT score can be used to indicate how good each model is. Typically, average pLDDT > 75 (or 0.75) indicate reliable prediction

###**pTM**: Predicted global folding accuracy. In other words, how accurately the whole protein would be folded. This ranges from 0 to 1, and pTM > 0.5 typically indicates correct folds.

## Important metrics to understand

###This is a lot of information! There are a several ew important pieces of information we need to understand.

###**TM-score**: Do you still remember pTM scores from AlphaFold? TM-align outputs TM-score as a measure of structural similarity. This score ranges from 0 (no similarity) to 1 (identical). TM-score > 0.5 typically indicates fold level similarity (the proteins are roughly folded in the same way). Greater scores indicate greater similarity. You will see two scores. That's because the score is normalized by the length of each protein.

###**alignment**: at the very bottom, you will notice the alignment between the proteins. This alignment comes from structural similarity, not from sequence similarity! Do you happen to see alignments for the pairs which you did not observe any detectable alignments?

###**Structural superposition**:On the bottom, you can see two selected structures superposed against one another. This is done by minimizing the distances between two backbones.

```python
#@title Perform all vs. all structural similarity searches
import subprocess

all_vs_all = f'all_vs_all'
if not os.path.exists(all_vs_all):
  !mkdir {all_vs_all}

#Run all_vs_all searches
seqs = ["05294","05295","05299","05300","05301","05302","05303","05305","05306","05309","05310","05311","05312","05314","05317","05318","05319","10057","10553","10554","10555","12302"]
folder = "PMB200B/structures"
k =  1
for i in range(0, len(seqs)):
  for j in range(i+1, len(seqs)):
    i_seq = seqs[i]
    j_seq = seqs[j]
    with open(f"{all_vs_all}/{i_seq}_vs_{j_seq}.tmalign.out", "w") as o:
      command = f"/content/TMalign {folder}/{i_seq}.pdb {folder}/{j_seq}.pdb"
      process = subprocess.Popen( command.split(" "), stdout = o)
      print(f"{k}: Procssed comparison between {i_seq} and {j_seq}")
      k += 1

```

```python
#@title Tabulate the output
with open(f'all_vs_all.summary.out', 'w') as o:
  for tmout in [ x for x in os.listdir(all_vs_all) if x.endswith("tmalign.out")]:
    full_tmout = f'{all_vs_all}/{tmout}'
    query = tmout.split('_')[0]
    hit   = tmout.split('vs_')[1].split('.')[0]

    for line in open(full_tmout, 'r'):
        if line.startswith("TM-score"):
              if "Chain_1" in line:
                tm1 = float(line.split()[1])
              elif "Chain_2" in line:
                tm2 = float(line.split()[1])

    o.write(f'{query}\t{hit}\t{max([tm1, tm2])}\n')

print('Structure 1  Structure 2  TM-score')
!head all_vs_all.summary.out
```

```python
#@title Visualize the output as a network
import networkx as nx
import matplotlib.pyplot as plt

similarity_cutoff = "0.5" #@param {type:"string"}

# Read the file
file_path = f'all_vs_all.summary.out'
protein_data = []

with open(file_path, 'r') as file:
    for line in file:
        line = line.strip().split()

        if float(line[-1]) > float(similarity_cutoff) and line[0] != line[1]:
          protein_data.append(line)

# Create a graph
G = nx.Graph()

# Add nodes and edges to the graph
for data in protein_data:
    protein1, protein2, score = data
    G.add_edge(protein1, protein2, similarity=float(score))

# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Layout for better visualization

# Draw nodes
node_colors_manual =  { '05294': 'pink', '05295': 'pink', '12302': 'pink', '10553': 'pink', '10554': 'pink',
                        '05299': 'lightgreen', '05300': 'lightgreen', '05301': 'lightgreen',
                        '05302': 'purple', '05303': 'purple', '10555': 'purple', '05305': 'purple', '05306': 'purple',
                        '05309': 'lightblue', '05310': 'lightblue', '05311': 'lightblue',
                        '05312': 'orange', '05314': 'orange', '10057': 'orange', '05317': 'orange', '05318': 'orange', '05319': 'orange'}

node_color = [ node_colors_manual.get(node) for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=200, node_color=node_color)

# Draw edges
nx.draw_networkx_edges(G, pos, width=0.5, edge_color='gray')

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8, font_color='black')


# Display the graph
plt.axis('off')
plt.title('Protein Structure Similarity Network Based On Structural Similarity')
plt.show()
```

## 3. Evaluate your hypothesis on homology

### It is possible for proteins with different ancestors adopt similar folds. **The protein space (or the number of protein folds) is much more limited than the sequene space (or different combinations of amino acids).**

### However, the probabily of four different homologous groups and five singletons adopting the same fold in the clustered genomic region is extemly low (I can confidently say it will be zero). In other words, this particular hypothesis is not evolutionary plasible.

### Remember many genes are created by **duplication events**. Pathogens are typically under very high evolutionary pressures for survival, which also places effectors under **strong selection**. It turns out the most plasible evolutionary explanation is that **this gene cluster originated from a single ancestor and underwent repeated duplication and divergence** (just like NLRs or many other expanded gene families across tree of life!) (Fig. 7). The divergence was simply so rapid and divergent that no sequence-based method could have detected the similarity of these homologs. With AlphaFold, however, we can now glance at such evolutionary aspects!

# Conclusion

##When you are dealting with evolution--whether it is homology, multiple sequence alignments, or phylogenetic tress, always remember that **we do not know true evolution**. We are simply **inferring (or predicting!) evolution based on sequences and structures**. There are always false positives (non-homologs detected to be homologous) and false negatives (homologs detected not to be homologous). Stay critical during data interpretation, and understand well the characteristics of your sequences and proteins!
